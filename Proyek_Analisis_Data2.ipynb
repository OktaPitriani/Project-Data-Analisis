### Cleaning Data

```python
missing_data = df.isnull().sum()
print(missing_data)
```

```python
df['dteday'] = pd.to_datetime(df['dteday'])
df['year'] = df['dteday'].dt.year
df['month'] = df['dteday'].dt.month
df['weekday'] = df['dteday'].dt.weekday
```

```python
df['cnt'].describe()

df = df[df['cnt'] < df['cnt'].quantile(0.99)]
```

```python
bins = [0, 10, 20, 30, 40]
labels = ['Very Cold', 'Cold', 'Warm', 'Hot']
df.loc[:, 'temp_cat'] = pd.cut(df['temp']*41, bins=bins, labels=labels)
```

```python
avg_rentals_by_day = df.groupby('workingday')['cnt'].mean()
print(avg_rentals_by_day)
```

```python
avg_rentals_by_weather = df.groupby('weathersit')['cnt'].mean()
print(avg_rentals_by_weather)
```

**Insight:**
- Difference between weekdays and non-weekdays: The average bicycle rental on weekdays (4,562) is slightly higher compared to non-weekdays (4,239).
- weathersit = 1 (clear/partly cloudy): This is the best weather condition for cycling, resulting in the highest average bike rentals (4,823). This suggests that people are more likely to use their bikes when the weather is clear or lightly cloudy.
  Clear weather (weathersit = 1) clearly supports higher bike usage. In this type of weather, many people use their bikes for both daily transportation and recreation.

## Conclusion

- Bike Rental Patterns:

  There is a significant difference in bike rental patterns between weekdays and weekends. Weekdays tend to have higher rental rates (average 4562) than weekends (average 4239), indicating that many people use bicycles for commuting.

  Weather also plays a significant role in the decision to rent a bike. The data shows that on days with good weather (weathersit = 1), the highest rental rates are recorded, while bad weather (weathersit = 4) shows the lowest rental rates.

- Weather Effects:

  The positive correlation between temperature (temp) and bike rental rates suggests that more rentals are made when the temperature is comfortable, while more extreme weather conditions (such as rain or snow) reduce rental rates.
